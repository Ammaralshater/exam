```python
import yfinance as yf
import matplotlib.pyplot as plt

tesla_data = yf.download("TSLA", start="2020-01-01", end="2023-01-01")
def make_graph(stock_data, title):
    plt.figure(figsize=(10, 6))
    
    plt.plot(stock_data['Close'], label='Tesla Stock Price', color='blue')
    
    plt.title(title, fontsize=16)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Stock Price (USD)', fontsize=14)

    plt.grid(True)
    plt.legend(loc='upper left')
    
   
    plt.show()

make_graph(tesla_data, "Tesla Stock Price Over Time (2020-2023)")
```
